```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

# Initialize Sentiment Analyzer
sia = SentimentIntensityAnalyzer()

# Analyze sentiment of news/twitter related to a particular stock
def analyze_sentiment(text):
    sentiment = sia.polarity_scores(text)
    return sentiment

# Example usage
news_sentiment = analyze_sentiment("Great news for Apple Inc., they just launched a new product!")
print("Sentiment:", news_sentiment)
```

```python
from whoosh.index import create_in
from whoosh.fields import Schema, TEXT, ID
from whoosh.qparser import QueryParser

# Create schema for search index
schema = Schema(title=TEXT(stored=True), content=TEXT(stored=True), path=ID(stored=True))

# Create index
index_dir = "/content/stock-market-analysis-data"
if not os.path.exists(index_dir):
    os.mkdir(index_dir)
ix = create_in(index_dir, schema)

# Add documents to the index
writer = ix.writer()
writer.add_document(title="Document 1", content="This is the content of document 1.")
writer.add_document(title="Document 2", content="This is the content of document 2.")
writer.commit()

# Perform search
with ix.searcher() as searcher:
    query = QueryParser("content", ix.schema).parse("content:search")
    results = searcher.search(query)
    for hit in results:
        print(hit)
```
